```python
df_final.groupby(['Aggregate rating','Rating color','Rating text'])
```

```python
df_final.groupby(['Aggregate rating','Rating color','Rating text']).size()
```

```python
df_rating = df_final.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
df_rating
```

# Observations 

1. The rating b/w 4.5 - 4.9 is excellent
2. The rating b/w 4.0 - 4.5 is very good
3. The rating b/w 3.5 - 3.9 is good
4. The rating b/w 2.5 - 3.4 is average
5. The rating b/2 1.8 - 2.4 is poor
6. Total of 2148 peoples are there who did not rated zomato and have rating of 0
7. According to the rating range color names are mention from which we can easily understand the rating flows

```python
import matplotlib
matplotlib.rcParams['figure.figsize'] = (12,6)
sns.barplot(x="Aggregate rating", y="Rating Count", data=df_rating)
```

```python
sns.barplot(x="Aggregate rating", y="Rating Count", data=df_rating, hue='Rating color')
```

```python
sns.barplot(x="Aggregate rating", y="Rating Count", data=df_rating, hue='Rating color',palette=['white','red','orange','yellow','green','green'])
```

# Observations:
   1. Not Rated Count is very high

```python
# Rating Color plot -

sns.countplot(x='Rating color',data=df_rating,palette=['blue','red','orange','yellow','green','green'])
```
